```python
import numpy as np
import pandas as pd

path0="/users/dten0001/Google Drive/proj_den/"

# This is my own library which you might not have yet
import automata.xio as xio

pd.options.display.max_rows = 999
```

```python
# Read in a fasta file and load it into a pandas dataframe
contents = xio.read_fasta(path0+"d1_e_5811_v4.fasta", delimiter="", preview=3)
d0 = pd.DataFrame(data=contents, columns=["header", "seq"])
```

```python
# add new columns: cdate, country ("geoloc"), and cyear
d0["cdate"] = d0.apply(lambda row: str(row["header"]).split("|")[-1], axis=1)
d0["geoloc"] = d0.apply(lambda row: str(row["header"]).split("|")[1], axis=1)
d0["cyear"] = d0.apply(lambda row: str(row["cdate"])[:4], axis=1)
```

```python
# Preview the first 5 columns of d0
d0.head()
```

```python
# Select only the Brunei rows
d1 = d0.loc[d0["geoloc"]=="Brunei"]
d1
```

```python
# Show the number of records by country and year
d_temp = d0.groupby(["geoloc", "cyear"]).size()
print(d_temp)
```

```python
# write out
d_temp.to_csv(path0+"dengue_dates.csv")
```
